## 1. Design and analyze a randomized proportion test
Power is chosen before observing outcomes; the result reports effect and uncertainty.


```python
from causal_toolkit.core import *
print('required per arm:',sample_size(.10,.02))
ab=proportion_effect(100,1000,130,1000)
print({k:(round(v,4) if isinstance(v,float) else tuple(round(x,4) for x in v)) for k,v in ab.items()})
```

## 2. Estimate observational counterfactuals
DiD uses change in a comparison group; synthetic control learns nonnegative donor weights.


```python
print('DiD:',did(10,15,9,11))
donors=np.array([[1,2,3],[2,2,2],[3,2,1]],float)
synth=synthetic_control(donors,np.array([1.5,2,2.5]),4)
print('weights:',synth['weights'].round(3),'pre-RMSE:',round(synth['pre_rmse'],4),'effect:',round(synth['effect'],3))
```

```python
assert ab['ci'][0]>0 and synth['pre_rmse']<.05
print('causal diagnostics passed')
```
