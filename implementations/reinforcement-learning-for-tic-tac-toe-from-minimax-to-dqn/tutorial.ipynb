## 1. Verify the game engine against minimax
The solved oracle supplies exact actions and catches terminal or illegal-move bugs.


```python
from tictactoe_rl.core import *
board=np.array([1,1,0,-1,-1,0,0,0,0]); print('winner:',winner(board),'minimax:',minimax(tuple(board),1))
assert minimax(tuple(board),1)[1]==2
```

## 2. Train a replay-based Q-network
For a fast deterministic tutorial, exact minimax transitions populate replay and the neural Q-function distills the oracle.


```python
cfg=DQNConfig(episodes=900); out=train(cfg)
print('replay:',out['replay_size'],'oracle action accuracy:',round(out['accuracy'],3),'checkpoints:',out['history'])
```

```python
assert out['accuracy']>.65
print('DQN evaluation passed')
```
