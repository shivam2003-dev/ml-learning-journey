```python
from image_gen.core import *
cfg=GeneratorConfig(); images,labels=make_data(cfg); model=train(cfg)
print('codebook:',model['codebook'].round(2),'reconstruction MSE:',round(model['reconstruction_mse'],5))
```

## 2. Autoregressively sample prompt-conditioned codes
The tiny categorical model learns position-wise code distributions for two text prompts.


```python
vertical=generate(model,'vertical stripe',cfg); horizontal=generate(model,'horizontal stripe',cfg)
print('orientation scores:',round(orientation_score(vertical),3),round(orientation_score(horizontal),3))
print(vertical.round(1))
```

```python
assert model['reconstruction_mse']<.01
assert orientation_score(vertical)>.35 and orientation_score(horizontal)<-.35
print('multimodal generation passed')
```
